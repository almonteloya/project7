```python
import numpy as np
from typing import List, Dict, Tuple, Union
from numpy.typing import ArrayLike
from nn.nn import NeuralNetwork
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
nn_ar = [{'input_dim': 64, 'output_dim': 16, 'activation' : 'Relu'}, {'input_dim': 16, 'output_dim': 64, 'activation': 'Relu'}]

lr = .5
seed = 2
batch_size =50
epochs = 10
loss_function = "mse"

#test.fit()
```

```python
## Creating a neural network
test = NeuralNetwork(nn_ar,lr, seed, batch_size,epochs,loss_function)
```

```python
## Loading and exploring the data
```

```python
digits = load_digits()
```

```python
import matplotlib.pyplot as plt
#plt.gray()
#plt.matshow(digits.images[7])
#plt.show()
```

```python
## Split the digits dataset into a train and a validation set
```

```python
X = digits.data
X_train, X_test, y_train, y_test = train_test_split(X, X, test_size=0.33, random_state=42)
```

```python
## Train your neural network!
```

```python
#test.fit(X_train,y_train,X_test,y_test)
```

```python
batch_train_L=[]
batch_validation_L=[]
```

```python
y_pred,cache = test.forward(X_train)
# Calculating loss using the according loss function
loss_train = test._loss_function(y_train,y_pred)
# Add the value to loss history record
batch_train_L.append(loss_train)
## Now we do the back grop
grad_dict = test.backprop(y_train, y_pred, cache)
test._update_params(grad_dict) ## Update using the derivates
final_pred, _ = test.predict(X_test) ## check the validation data 
# Validation pass
loss_val = test._loss_function(y_test, final_pred)
batch_validation_L.append(loss_val)
```

```python
num_batches = int(X_train.shape[0]/(test._batch_size)) + 1
X_batch = np.array_split(X_train, num_batches)
y_batch = np.array_split(y_train, num_batches)
```
